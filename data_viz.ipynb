```python
import utils
import matplotlib.pyplot as plt
import numpy as np

datafile = 'preprocessed_data.csv'

start = 2010
train_X, all_y, X_mean, X_std, y_mean, y_std, _ = utils.get_data(start, 2018, datafile)
# y will be 2012 - 2019
```

```python
all_y = all_y.flatten()
plt.rcParams['font.size'] = '16'
plt.figure(figsize=(12, 4))
plt.plot(all_y, color="red")
year_labels = [i for i in range(2012, 2021)]
weeks = [i for i in range(len(all_y))]
locs, labels = plt.xticks(weeks[::52], year_labels)
plt.axvspan(0, 6*52, alpha=0.3, color='yellow', label="train labels")
plt.axvspan(7*52, 8*52, alpha=0.3, color='blue', label="test labels")
plt.legend()
plt.title("Dataset Label Visualization: 2012-2019 Flu Seasons")
plt.ylabel("Total US Influenza Cases")
plt.xlabel("Flu Season (begins week 40)")
```
